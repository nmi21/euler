Find the sum of multiples of 3 or 5 below 1000.

```python
nick_list = []

for num in range(1, 999 + 1):
    
    if num % 3 == 0 or num % 5 == 0:
        nick_list.append(num)
```

```python
sum(nick_list)
```

```python
my_sum = 0

for num in range(1, 1000):
    
    if num % 3 == 0 or num % 5 == 0:
        my_sum += num
        print(f"{num=}")
        print(f"{my_sum=}")
        print()
        
my_sum
```
