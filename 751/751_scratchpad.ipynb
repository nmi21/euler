```python
import math
from decimal import *
```

```python
c = getcontext()
c.traps[FloatOperation] = False
c.prec = 25
c
```

```python
def next_vals(b):
    
    next_b = Decimal(int(b)) * (b - Decimal(int(b)) + 1)
    next_a = Decimal(int(next_b))
    
    return next_a, next_b
```

```python
def tau_generator(theta):
    
    b1 = theta
    a1 = Decimal(int(b1))

    pair_list = [(a1, b1)]

    for i in range(25):
        a, b = next_vals(pair_list[-1][1])
        pair_list.append((a, b))

    # generate tau
    t = [str(x[0]) for x in pair_list]
    t.insert(1, ".")
    tau = "".join(t)
    tau = Decimal(tau[:26])

    return tau
```

```python
def delta(theta, tau):
    return tau - theta
```

```python
theta = Decimal('2.22356')
```

```python
tau = tau_generator(theta)
tau
```

```python
d = delta(theta, tau)
print(d)

if d > 0:
    print("add to theta")
elif d < 0:
    print("subtract from theta")
```

```python
theta = Decimal(2)
inc = Decimal('.1')

d = 1

while d > 0:
    theta += inc
    tau = tau_generator(theta)
    d = delta(theta, tau)
    
    if d < 0:
        theta -= inc
        inc /= 10
        d = 1
    
    print(theta, tau, d)
```
