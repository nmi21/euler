```python
import math
import itertools
```

```python
fibonaccis = [0, 1, 1]

while len(fibonaccis) < 123:
    fibonaccis.append(fibonaccis[-1] + fibonaccis[-2])
```

```python
fibonaccis
```

```python
fibs_mod = [x % 30 for x in fibonaccis]
```

```python
fibs_mod[120]
```

```python
fib_120 = fibonaccis[120]
fib_120
```

```python
prime_factors = []

num = fib_120
for i in range(2, int(math.sqrt(fib_120))):
    while num % i == 0:
        prime_factors.append(i)
        num = num // i
    if i > num:
        break
```

```python
prime_factors
```

```python
prime_counts = {}
for factor in prime_factors:
    prime_counts[factor] = prime_counts.get(factor, 0) + 1
```

```python
exponents = [[i for i in range(count + 1)] for count in prime_counts.values()]
```

```python
prime_combos = itertools.product(*exponents)

for pc in prime_combos:
    print(pc)
```

```python
factors = []

for combo in prime_combos:
    factor = 1
    for p, e in zip(prime_counts.keys(), combo):
        factor *= p ** e
    factors.append(factor)

factors.sort()
```

```python
factors = [x for x in factors if x <= 1_000_000_000]
```

```python
len(factors)
```

```python
res = []
for f in factors:
    fibs_mod = [x % f for x in fibonaccis]
    if fibs_mod[-3:] == [0, 1, 1]:
        res.append(f)       
        
        
remove = []
for r in res:
    temp = [x % r for x in fibonaccis]
    temp = temp[3:]
    temp = temp[:-3]
    
    for i in range(len(temp)):
        if temp[i] == 0 and temp[i+1: i+3] == [1, 1]:
            remove.append(r)
            break
            
for r in remove:
    res.remove(r)
```
