2^7 = 128 is the first power of two whose leading digits are "12". The next power of two whose leading digits are "12" is 2^80.

Define p(L,n) to be the n'th-smallest value of j such that the base 10 representation of 2^j begins with the digits of L.

So p(12,1) = 7 and p(12,2) = 80.

You are given that p(123,45) = 12710.

Find p(123,678910).

2^j = 10^n

j * log10(2) = n * log10(10)

j * log10(2) = n

```python
from math import floor, log10

print(log10(12))
print(log10(13))
print(log10(120))
print(log10(130))
print(log10(1200))
print(log10(1300))
```

```python
def p(L, n):
    
    k_log10_2 = log10(2)
    
    lower = log10(L) - floor(log10(L))
    upper = log10(L + 1) - floor(log10(L + 1))
    
    counter = 0
    j = 1
    while True:
        x = j * k_log10_2
        x -= floor(x)
        if lower <= x <= upper:
            counter += 1
            
            if counter % 1000 == 0:
                print(j, counter)
            
            if counter == n:
                break
            
        
        j += 1
        
        
        
    return j
```
