```python
def partitions(n):
    # base case of recursion: zero is the sum of the empty list
    if n == 0:
        yield []
        return

    # modify partitions of n-1 to form partitions of n
    for p in partitions(n-1):
        yield [1] + p
        if p and (len(p) < 2 or p[1] > p[0]):
            yield [p[0] + 1] + p[1:]
```

```python
def can_make_palindrome(partition_tuple):
    
    counter = 0
    odd_num = None
    
    # see how many items have quantities
    for p in set(partition_tuple):
        if partition_tuple.count(p) % 2 == 1:
            counter += 1
            odd_num = p
            
            if counter > 1:
                return False, odd_num
            
    # even length tuples must have even quantities of nums
    if len(partition_tuple) % 2 == 0 and counter == 0:
        return True, odd_num
    
    # odd length tuples must have 1x odd quantities of nums
    if len(partition_tuple) % 2 == 1 and counter == 1:
        return True, odd_num
    
    return False, odd_num
```

```python
def qty_palindromes(partition_tuple, odd_num):
    """
    Returns quantity of palindromes that can be made from a valid partition tuple
    """
    
    temp = partition_tuple
    
    # first need to remove the odd_num if it exists
    if odd_num:
        temp = list(temp)
        temp.remove(odd_num)
        
    uniques = set(temp)
    uniques_length = len(uniques)
    
    if uniques_length < 1:
        return 0
    
    n = len(temp) // 2
    k = uniques_length - 1
    return math.perm(n, k)
    
```

```python
def num_twopals(num):
    
    # generate all partitions with 2 in them
    parts = [tuple(x) for x in list(partitions(num)) if 2 in x]
    
    # start a counter
    res = 0
    
    for part in parts:
        possible, odd_qty = can_make_palindrome(part)
        
        if possible:
            res += qty_palindromes(part, odd_qty)
            
    return res
```

```python
def count_palindromes(num_tup):
    
    perms = itertools.permutations(num_tup)
    
    perm_set = set()
    
    for p in perms:
        perm_set.add(p)
    
    palins = []
    
    for p in perm_set:
        if is_tuple_palindrome(p):
            palins.append(p)
            
    return len(set(palins))
```
