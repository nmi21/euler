```python
# concatenate all of the strings together
# x.join(y) will use 'x' as a separator to join all elements of y
# e.g. "a".join([1, 2, 3]) will return "1a2a3"
# you can use an empty string to join without any characters between
big_num = "".join(lines)

big_num
```

```python
def get_string_prod(number_string):
    """
    Takes in a string of any length and gives the digital product
    """
    
    prod = 1
    for d in number_string:
        prod *= int(d)
    
    return prod
```

```python
def get_max_prod(number_string, num_digits):
    """
    Takes in a number_string and gives the maximum product from consecutive digits
    Quantity of digits is given by num_digits
    """
    
    # initialize a maximum product
    max_prod = 0
    
    # search through strings of length num_digits
    # thus need to stop early enough else will run into an index error
    end = len(number_string) - num_digits
    for i in range(end + 1):
        
        # grab the substring
        sub_str = number_string[i : (i + num_digits)]
        
        # calculate the product of the substring
        prod = get_string_prod(sub_str)
        
        # compare the product to the stored max value
        if prod > max_prod:
            max_prod = prod
            
    return max_prod
```
