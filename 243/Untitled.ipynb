```python
import math
from fractions import Fraction
```

```python
def build_sieve(num):
    sieve = [True] * (num + 1)
    sieve[0] = False
    
    for i in range(2, int(math.sqrt(num) + 2)):
        if sieve[i] and num % i == 0:
            for j in range(i, num + 1, i):
                sieve[j] = False
                
    return sieve
```

```python
def get_resiliance(num):
    sieve = build_sieve(num)
    return Fraction(sieve.count(True), len(sieve) - 2)
```

```python
lim = Fraction(15499, 94744)
# lim = Fraction(4, 10)

n0 = 30030
num = n0
while True:
    print(f"{num=}")
    
    if get_resiliance(num) < lim:
        break
    num += n0
    
    
print()
print("Found the answer!")
print(num)
```
